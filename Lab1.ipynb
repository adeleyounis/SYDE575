2. Image Quality Measures

```python
def PSNR(f: np.ndarray , g: np.ndarray) -> float:
    """
    Params:
        f: Reference Image
        g: Test Image
    Returns:
        PSNR_out: Peak Signal to Noise ratio output value
    """
    # Set MAX_f
    MAX_f = np.max(f)

    # convert to float 64's
    float_f = f.astype(np.float64)
    float_g = g.astype(np.float64)

    # calculate the MSE
    MSE = np.sum((float_f - float_g) ** 2) / (float_f.shape[0] * float_g.shape[1])

    # get PSNR
    PSNR_out = 10*np.log10((MAX_f**2)/MSE)
    return PSNR_out
```

3. Digital Zooming

```python
# Prepare Images
lena = rgb2gray(lena)*255

lena_ds = skimage.transform.resize(lena, (lena.shape[0]/4, lena.shape[1]/4))

cameraman_ds = skimage.transform.resize(cameraman, (cameraman.shape[0]/4, cameraman.shape[1]/4))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].imshow(lena_ds, cmap='gray')
axes[0].set_title("Lena")
axes[0].axis("on")

axes[1].imshow(cameraman_ds, cmap='gray')
axes[1].set_title("Cameraman")
axes[1].axis("on")
```

```python
# Nearest Neighbour (resize order = 0)
lena_NN = skimage.transform.resize(lena_ds, (lena_ds.shape[0]*4, lena_ds.shape[1]*4), order=0)
cameraman_NN = skimage.transform.resize(cameraman_ds, (cameraman_ds.shape[0]*4, cameraman_ds.shape[1]*4), order=0)

# Bilinear Interpolation (resize order = 1)
lena_BLI = skimage.transform.resize(lena_ds, (lena_ds.shape[0]*4, lena_ds.shape[1]*4), order=1)
cameraman_BLI = skimage.transform.resize(cameraman_ds, (cameraman_ds.shape[0]*4, cameraman_ds.shape[1]*4), order=1)

# Bicubic Interpolation (resize order = 3)
lena_BCI = skimage.transform.resize(lena_ds, (lena_ds.shape[0]*4, lena_ds.shape[1]*4), order=3)
cameraman_BCI = skimage.transform.resize(cameraman_ds, (cameraman_ds.shape[0]*4, cameraman_ds.shape[1]*4), order=3)
```

```python
# Plot
fig, axes = plt.subplots(2, 4, figsize=(12,8))

axes[0][0].imshow(lena_NN, cmap='gray')
axes[0][0].set_title("Lena Nearest Neighbour")

axes[1][0].imshow(cameraman_NN, cmap='gray')
axes[1][0].set_title("Cameraman Nearest Neighbour")

axes[0][1].imshow(lena_BLI, cmap='gray')
axes[0][1].set_title("Lena Bilinear Interpolation")

axes[1][1].imshow(cameraman_BLI, cmap='gray')
axes[1][1].set_title("Cameraman Bilinear Interpoloation")

axes[0][2].imshow(lena_BCI, cmap='gray')
axes[0][2].set_title("Lena Bicubic Interpolation")

axes[1][2].imshow(cameraman_BCI, cmap='gray')
axes[1][2].set_title("Cameraman Bicubic Interpolation")

axes[0][3].imshow(lena, cmap='gray')
axes[0][3].set_title("Lena Original")

axes[1][3].imshow(cameraman, cmap='gray')
axes[1][3].set_title("Cameraman Original")


fig.tight_layout()
```

```python
# Calculate PSNR
lena_PSNR_NN = PSNR(lena, lena_NN)
lena_PSNR_BLI = PSNR(lena, lena_BLI)
lena_PSNR_BCI = PSNR(lena, lena_BCI)

print(f"Lena PSNR's:")
print(f'Nearest Neighbour: {lena_PSNR_NN}, Bilinear Interpolation {lena_PSNR_BLI}, Bicubic Interpolation {lena_PSNR_BCI}')

cameraman_PSNR_NN = PSNR(cameraman, cameraman_NN)
cameraman_PSNR_BLI = PSNR(cameraman, cameraman_BLI)
cameraman_PSNR_BCI = PSNR(cameraman, cameraman_BCI)

print(f"Cameraman PSNR's:")
print(f'Nearest Neighbour: {cameraman_PSNR_NN}, Bilinear Interpolation {cameraman_PSNR_BLI}, Bicubic Interpolation {cameraman_PSNR_BCI}')

```

4. Point Operations for Image Enhancement

**6. Explain what the histogram of an image represents. Why is it useful?**

The histogram of an image represents the quantity of pixels that have each gray level across the entire image intensity range. The histogram is useful because it allows for quick comparison of images in terms of pixel distribution, which can reveal information about important qualities like contrast, where a correcting transformation or enhancement can then be selected in order to improve image quality.

**7. Describe how the histogram looks like in the context of intensity distribution. What does the histogram
say about the image?**

The image has a high number of low-intensity pixels, and a comparatively very low number of medium and high-intensity pixels. Based on this, the image is mostly comprised of darker areas, but maintains high contrast because of the low (but relatively even) spread of medium/high pixels.

```python
tire = imread('tire.tif').astype(np.float64) / 255.0
tire.flatten()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].imshow(tire, cmap='gray')
axes[0].set_title("Original Tire")
axes[1].hist(tire)
axes[1].set_title("Original Tire Histogram")

plt.show()
```


**8. Describe how the histogram looks like in the context of intensity distribution. How does it differ from
the histogram of the original image? Why?**

The negative histogram has a large number of high-intensity pixels, and a small, but even spread of medium/low-intensity pixels. This differs from the original histogram in that the high-intensity and low-intensity ranges have been swapped- this histogram is a horizontally reflected version of the original. This difference is because a negative image transformation maps high intensities to low intensities, and vice versa, leading to an inversion of the distribution of pixels, and therefore an inversion of the histogram itself.

```python
# Perform negative transform
from skimage import util
negative = util.invert(tire)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].imshow(negative, cmap='gray')
axes[0].set_title("Negative Tire")
axes[1].hist(negative)
axes[1].set_title("Negative Tire histogram")

plt.show()
```

```python
# Apply both power law transformations and compare
y1 = 0.5
y2 = 1.3
t_gamma1 = tire**y1
t_gamma2 = tire**y2
t_gamma1.flatten()
t_gamma2.flatten()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0][0].imshow(t_gamma1, cmap='gray')
axes[0][0].set_title("=0.5")
axes[0][1].hist(t_gamma1)
axes[0][1].set_title("=0.5")
axes[1][0].imshow(t_gamma2, cmap='gray')
axes[1][0].set_title("=1.3")
axes[1][1].hist(t_gamma2)
axes[1][1].set_title("=1.3")

plt.show()

```

**9. Describe the appearance of the transformed images. Why do they appear this way?**

The image transformed with =0.5 appears to have more detail in regions that were previously darker in the original image, like the tire itself, whereas the image transformed with =1.3 appears to lose detail compared to the original, with most regions becoming darker/underexposed. The first image appears this way because using an exponent of  < 1 effectively stretches the dark range of the image, leading to a brighter overall result, and more discernable features in the stretched region. Using  > 1 does the opposite, compressing the dark range of the image, leading to a darker overall result, and less detail in the compressed region.

**10. Describe how each of the histograms look like in the context of intensity distribution. Why do they

The histogram of the =0.5 transformed image has a mostly even distribution of intensities, with slightly more dark pixels. The =1.3 histogram has a much higher number of low-intensity pixels, and a very small (but even) spread of medium/high intensity pixels. The =0.5 histogram appears this way because the lower range of the image has been stretched and the higher range compressed, leading to a relatively even distribution of intensities due to the content of the original image histogram. The =1.3 histogram appears this way because the upper range of the image has been stretched, and the lower range compressed, leading to an even higher number of dark pixels when compared to the original.

**11. Compared with the original image, which of the transforms should you use to enhance the image?
Why?**

The =0.5 gamma correction should be used to enhance the image, as the result of the transformation has greater detail, and does not appear undersaturated when compared with the original. The original image has a high number of dark pixels and a low number of medium/high pixels. In order to get a more even distribution, the lower range must be stretched, and the higher range compressed, which is achieved using =0.5.
```python
# Histogram Equalization
from skimage import exposure

equalized = exposure.equalize_hist(tire)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].imshow(equalized, cmap='gray')
axes[0].set_title("Equalized Tire")
axes[1].hist(equalized)
axes[1].set_title("Equalized Tire histogram")
```

***12. Describe the appearance of the equalized image.***

The equalized image appears lighter than the original dark tire image. The overall image contrast is higher, making details in the tire, shadow, and pavement more visible. The original image looked a lot more flat in terms of image contrast, whereas the equalized image looks a lot more balanced, as the light and darker sections of the image are blended.

***13. Describe how the histogram looks like in the context of its intensity distribution. Why does it look like this? What does the histogram say about the equalized image?***

The histogram of the equalized image is spread out across the full intensity region, rather than mostly in the darker range as is in the original histogram. Although the histogram is not completely uniform, it shows that the pixel intensities were redistributed evenly across all the intensities. This redistribution aligns with the overall lighter image we see in the equalized image, as the equalized image stretches out across all the intensities, pulling in some of the darker shades into the lighter shades. This allows us to see new details in the image that were originally hidden in the dark intensities.
